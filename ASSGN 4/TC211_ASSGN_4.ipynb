```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = pd.read_csv("/home/mmcoe/Downloads/HousingData.csv")
```

```python
a
```

```python
a.isnull().sum()
```

```python
a.fillna(method='ffill', inplace = True)
```

```python
a.isnull().sum()
```

```python
# Here, we find the correlation between the variables and display correlation using heatmap.
corr = a.corr()
corr
```

```python
plt.subplots(figsize=(10,9))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# Based on the correlation values we will decide 'x' and 'y' values.
# x is independent variable and y is dependent variable.
# x will be from column 0 to 12 and y will last column.
x = a.iloc[:,0:13]
y = a.iloc[:,13]
print(x.shape,y.shape)
```

```python
# Here,we split the data. Based on 70-30 or 80-20 i.e (70= train_data and 30 = test_data)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# Here, we create LinearRegression object.
# From Linear Regression, we get the slope and intercept value.
# Using 'Slope' and 'Intercept' value we can predict the 'y' values i.e (y=mx+c)
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
print(regression.intercept_ ,regression.coef_)
# regression.intercept_ -->  It is the intercept value
# regression.coef_ -->  It is a slope value
```

```python
from sklearn.metrics import r2_score
y_predict_train=regression.predict(x_train)
r2_sc_train = r2_score(y_train,y_predict_train)
r2_sc_train
```

```python
y_predict_test = regression.predict(x_test)
r2_sc_test = r2_score(y_test,y_predict_test)
r2_sc_test
```

```python
from sklearn.metrics import mean_squared_error
rmse = (np.sqrt(mean_squared_error(y_train, y_predict_train)))
rmse
```
